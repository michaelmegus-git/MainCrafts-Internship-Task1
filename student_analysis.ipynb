# Data Analysis Project

**Load Data**

```python
import pandas as pd

df=pd.read_csv("student_dataset.csv")
df.head()
```

**Clean Data**

- check missing values

```python
df.isnull().sum()
```

- Remove Duplicates

```python
df.drop_duplicates()
```

- Data shape

```python
df.shape
```

- Dataset datatypes

```python
df.dtypes
```

**Analyze Data**

- Average final grade (G3)?

```python
df["G3"].mean()
```

- How many students scored above 15?

```python
(df["G3"]>15).sum()
```

- Is study time correlated with performance?

```python
df["studytime"].corr(df["G3"])
```

- Which gender performs better on average?

```python
avg_score=df.groupby(df["sex"])["G3"].mean()
print(avg_score)
```

**Visualization**

- Histogram of grades.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["G3"],bins=10,kde=True)
plt.xlabel("Grades")
plt.title("Distribution of final grades")
plt.show()
```

- Scatterplot: study time vs grades

```python
x_data=df["studytime"]
y_data=df["G3"]
plt.scatter(x_data,y_data)
plt.xlabel("study time")
plt.ylabel("Grades")
plt.title("Study time vs Grades")
plt.show()
```

- Bar chart: male vs female average score

```python
avg_score.plot(kind="bar")
plt.xlabel("gender")
plt.ylabel("average grades")
plt.title(" male vs female average score")
plt.show()
```
